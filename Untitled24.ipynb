```python
df.head(10)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
print(df.describe(include='all'))
```

```python
missing = df.isnull().mean().sort_values(ascending=False).head(10)
print(missing)
```

```python
trader_metrics = df.groupby('Account').agg(
    total_pnl=('Closed PnL', 'sum'),
    trade_count=('Trade ID', 'count')
).reset_index()
print(trader_metrics.head())
```

```python
stable_traders = trader_metrics[trader_metrics['trade_count'] >= 10]
print(stable_traders.head())
```

```python
if 'date' in df.columns:
    daily = df.groupby('date').agg(
        daily_ret=('pnl', 'mean'),
        avg_sent=('sentiment', 'mean')
    ).reset_index()

    fig, ax1 = plt.subplots()
    ax1.plot(daily['date'], daily['daily_ret'], color='blue')
    ax1.set_ylabel('Daily Return', color='blue')

    ax2 = ax1.twinx()
    ax2.plot(daily['date'], daily['avg_sent'], color='orange')
    ax2.set_ylabel('Avg Sentiment', color='orange')
    plt.show()
```

```python
sns.scatterplot(data=trader_metrics, x='trade_count', y='total_pnl')
plt.title('Trader Avg count vs Total PnL')
plt.show()
```

```python
features = stable_traders[['total_pnl', 'trade_count']]
kmeans = KMeans(n_clusters=3, random_state=0)
stable_traders['cluster'] = kmeans.fit_predict(features)

print(stable_traders['cluster'].value_counts())
```

```python
sns.scatterplot(
    data=stable_traders,
    x='trade_count', y='total_pnl', hue='cluster', palette='tab10'
)
plt.title('Trader Clusters')
plt.show()
```
